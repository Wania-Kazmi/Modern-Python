```python
class StudentCard():
    def __init__(self, roll_no: int, name:str, age:int) -> None:
        if age < 18 or age > 60:
            raise StudentAgeError("You're not eligible for this program!")
        self.age = age
        self.roll_no = roll_no
        self.name = name    

#custon error class
class StudentAgeError(Exception):   #means its parent class is Exception that generate error
    pass

student1 = StudentCard(1, "Wania", 20)
print(student1.roll_no, student1.name, student1.age)
```

### Now we want that we write a code and then read the whole code 

## Difference between below 2 codes 
### Code 1:
* print(file.read()): This line reads the entire content of the file and prints it to the console.

* print(file.write("I am python lover \n")): This line writes the string "I am python lover" followed by a newline character to the file. The write method returns the number of characters written, which will be printed by the print statement.

* print("After", file.read()): This line attempts to read the content of the file again after writing to it. However, this might not show the updated content because the file cursor is at the end of the file after the write operation. So, it's likely to print an empty string or part of the content.

```python
with open("./abc1.txt", 'r+') as file:
    print(file.read())
    print(file.write("I am python lover \n"))
    display("After",file.read()) #no result because file cursor is at the end
```

```python
with open("./abc1.txt", 'r+') as file:
    print(file.read())
    print(file.write("I am python lover \n"))
    file.seek(0)
    print("After",file.read())
```

```python
from typing import List
import PyPDF2

def read_pdf(file_path: str) -> List[str]:
    with open(file_path, 'rb') as file: #This b in rb stands for binary mode
        reader : PyPDF2.PdfReader = PyPDF2.PdfReader(file)
        text_content : List[str] = [page.extract_text() for page in reader.pages]
        return text_content

pages : list[str] = read_pdf("./mypdf.pdf")
print(pages)
```
