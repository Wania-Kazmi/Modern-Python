## When we apply slicing on series

```python
s1 : pd.Series = pd.Series([1,2,3,4,5])
display(s1)
print("After slicing ====")
s1[1:4]
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5])
display(s1)
print("After slicing ====")
s1[:2]
```

```python
def grading(sc:int,phy:int,math:int,chm:int,cs:int)->tuple[float,int,float,str]:
    obtained:float = sc + phy + math + chm + cs
    total:int = 500
    per: float = obtained/total * 100
    grade = ""
    if per >= 90:
        grade = "A"
    elif per >= 80:
        grade = "B"
    elif per >= 70:
        grade = "C"
    elif per >= 60:
        grade = "D"
    else:
        grade = "F"
    return obtained, total, per, grade
    
```

```python
combine['Student Data'] = combine[["Science","Physics","Math","Chemistry","CS"]].apply(lambda x: grading(*x), axis = 1) #without expansion
combine.head()
```

```python
combine[["N_obtained","N_total","N_percentage","N_grade"]] = combine[["Science","Physics","Math","Chemistry","CS"]].apply(lambda x: grading(*x), axis = 1, result_type= "expand")
combine.head()
```
