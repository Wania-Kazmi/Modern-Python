### Multiple input one output

### Multiple input multiple output

```python
def fn_obtained(science : int, physics : int, chemistry : int, math : int, computer_science : int)->tuple[int,int,float,str]:
    obtained : int = science + physics + chemistry + math + computer_science
    total:int = 500
    per : float = obtained / total * 100
    grade : str = ""
    if per >= 80:
        grade = 'A+'
    elif per >= 70:
        grade = 'A'
    elif per >= 60:
        grade = 'B'
    elif per >= 50:
        grade = 'C'
    elif per >= 40:
        grade = 'D'
    elif per >= 33:
        grade = 'E'
    else:
        grade = 'F'
    return obtained, total, per, grade

df1[["Obtained","Total","Percentage","Grade"]] = df1[["Science","Physics","Chemistry","Math","Computer Science"]].apply(lambda x: fn_obtained(*x), axis = 1, result_type = 'expand')
df1
```
